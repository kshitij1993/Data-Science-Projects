***Therefore a dictionary 'column_names' has been created consisting of 'old names : new names' in a key:value pair fashion***

***Thereafter,we make use of 'rename()' function of os library to facilitate the renaming of columns.Column 'DBN' is renamed to 'Database_Number'.Then 'SCHOOL NAME' has been renamed to 'School_Name' and so on.***

*** .dropna(thresh=4) : it means that interpreter will keep only those rows having atleast 4 non-NaN values***

### What is Data Munging?

**While our investigation of the data, we found a couple of issues in the data , which should be eliminated before the information is prepared for a decent model. This activity is normally know as "Data Munging".**

### Potential Problems with our Dataset

- Data types are not of the same types,example we have numeric values alongwith string values(for example 's') in all the columns.
- Data comprised of inconsistent values which we need to handle for better analysis of information

### Possible solutions to work with inconsistent data

- Ignoring rows with missing values 
- Fill in missing value
- Correcting erroneous values
- Standardizing the values
- Remove duplicates, if any (in this case not) 

### Solution Applied for this DataSet : Fill in missing value

** Explanation :**

***The columns are renamed as per the user readability.Now we need to work on inconsistent values present in our data***

**Steps followed in below code :**
- 1) Replace string value 's' with NaN of Numpy library
- 2) Converting data types of all the columns to numeric using 'to_numeric' of Pandas library
- 3) Calculating the mean values of all the columns so that we can assing it in place of inconsistent values of respective columns

```python
result_data_csv.replace('s',np.NaN)#replace 's' with NaN
result_data_csv["Applicants"]=pd.to_numeric(result_data_csv["Applicants"], errors="coerce")#converts data type column Applicant to numeric 
result_data_csv["Avg_CR_score"]=pd.to_numeric(result_data_csv["Avg_CR_score"], errors="coerce")#converts data type column Avg_CR_score to numeric 
result_data_csv["Avg Math Score"]=pd.to_numeric(result_data_csv["Avg Math Score"], errors="coerce")#converts data type column Avg Math Score to numeric
result_data_csv["Avg Writing Score"]=pd.to_numeric(result_data_csv["Avg Writing Score"], errors="coerce")#converts data type column Avg Writing Score to numeric
 
print("Applicants Mean:",int(result_data_csv["Applicants"].mean()))#type casted mean for column Applicants to an integer value
print('\n')
print("Avg CR score Mean :",result_data_csv["Avg_CR_score"].mean())
print('\n')
print("Avg Math Score :",result_data_csv["Avg Math Score"].mean())
print('\n')
print("Avg Writing Score :",result_data_csv["Avg Writing Score"].mean())
```

**Explanation :**

** What is fillna() ?**

***fillna() : The fillna function can “fill in” NA values with non-null data ***

***In the below code block we have made use of fillna() for Pandas to fill in NaN values with non null data i.e. mean (which we have computed in the above cell) for that respective column.All the NaN values are handled in a same way.***

***We can see that NaN values are now replaced by mean in the below output.***


```python
result_data_csv["Applicants"].fillna(int(result_data_csv["Applicants"].mean()), inplace=True) #fills integer value of mean of Applicants column in place of NaN 
result_data_csv["Avg_CR_score"].fillna(int(result_data_csv["Avg_CR_score"].mean()), inplace=True)#fills integer value of mean of Avg_CR_score column in place of NaN
result_data_csv["Avg Math Score"].fillna(int(result_data_csv["Avg Math Score"].mean()), inplace=True)#fills integer value of mean of Avg Math Score column in place of NaN
result_data_csv["Avg Writing Score"].fillna(int(result_data_csv["Avg Writing Score"].mean()), inplace=True)#fills integer value of mean of Avg Writing Score column in place of NaN
 
result_data_csv #displays the value of dataframe
```

**Explanation: **

*** Below code block makes use of 'pd.melt()' function on dataframe.***

***Here melt() has two types of variables :***
- Identifier variables
- Measured variables

*** In this case,identifier variables are columns 'SNo' and 'Database_Number'***

***Measured variables are 'Avg_CR_score','Avg Math Score' and 'Avg Writing Score'.***

***The main objective of below code block is to take observations that are spread across columns (Avg_CR_score, Avg Math Score,Avg Writing Score) and melting them down into one column(variable) with multiple rows.Identifier variables however remain unaffected with their observations.***

## Data Visualisation :

***The next important thing to perform after Data cleaning is Data Visualisation.***

***Since the data is all cleaned up,we are all set to plot the graphs of different cases so as to facilitate the ease to decipher the information present in Dataframe.***

***We make use of few basic types of graphs for visualisation of information.***

### 1) Factorplot()
** Syntax: **

**Factorplot(x,y,data=None,kind={point,bar,violin},size) **
- x : Column to be displayed on x-axis
- y : Column to be displayed on y-axis
- data : Dataset from which we plot the graphs('result_data_csv' in this case)
- kind : kind specifies the type of graph to be displayed
- size : defines the dimensions of graph

** In the below code block we have plotted a factorplot of 'Applicants' vs 'School_Name' of such cases where number of Applicants is between 200 and 300 **

### Conclusion :
**We can conclude that there are following schools having applicants in the range of 200 and 300**

```python
sns.factorplot(x='Applicants',y='School_Name',data=result_data_csv.query('200<Applicants<300'),kind='bar',size=8)
sns.despine(left=True) #removes extra axis from left
```

### 2) Regplot()

** Syntax :**

**Regplot(x,y,data=None) **
- x : Column to be displayed on x-axis
- y : Column to be displayed on y-axis
- data : Dataset from which we plot the graphs('result_data_csv' in this case) and result_data_csv.query specifies the condition for which the graph is displayed
- scatter : is set to 'True' if we want the graph to be plotted in scatter points
- label : displays the text within the graph for better evalutaion of information to end-user


** In the below code block we have plotted a regplot of 'Avg_CR_score' vs 'Applicants' of such cases where number of Applicants is greater than 300**

### Conclusion:
**As visible from the plot,we can conclude that a maximum of Applicants have an average score in the range 350 and 400.Therefore,various conclusions can be drawn from this analysis on different levels.**

```python
g=sns.regplot(x='Avg_CR_score',y='Applicants',data=result_data_csv.query('300>Applicants'), scatter= True,label='Applicants')
plt.legend() # to display text within the plot
sns.despine(left=True) ##removes extra axis from left
```

### 3) Kernel Density plot

** Syntax : **

** kdeplot(data,shade,color) **

- data : Input dataframe
- shade : It represents a boolean value which is used to decide whether a shade is required or not
- color : It specifies the color of the graph

### Conclusion:
** Below plot helps us to identify the peak value as the trendline reaches the top for average score of approximate 370.**

```python
ax=sns.kdeplot(result_data_csv.Avg_CR_score.dropna(),shade=True,color='g')
```

### 4) Barplot()

** Syntax : **

**barplot(x,y,data,color) ** 
- x : Column to be displayed on x-axis
- y : Column to be displayed on y-axis
- data : DataFrame which is used a reference for plotting a distplot
- color : displays the bars in color specified in 'color' paramater

### Conclusion:

**We see that there are around six school having number of Applicants greater than 800.**

```python
fig, ax = plt.subplots(figsize=(25, 6)) #creates a figure with size 2500 by 600 pixels
sns.barplot(x='School_Name', y='Applicants', data=result_data_csv.query('800<Applicants'),
            color='k', ax=ax) #'School_Name' is on X-axis and 'Applicants' is on Y-axis.
                            #If number of Applicants is greater than 800 then barplot plots the graph as below
sns.despine() #Removes the top and right spines from plot.
```
