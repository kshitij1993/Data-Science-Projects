** Mathematical operations on Arrays**


**Note : Size of arrays should be same for performing Mathematical operations **

** 1) Addition **

```python
arr1= np.array([1,2,3,4,5], float)
arr2 = np.array([5,2,6,7,8], float)
sum=arr1+arr2
print("Sum of arr1 and arr2 :",sum)
```

Note : This error is deliberately kept in the document for the reader to get overview of errors occuring due to size mismatches of two arrays

```python
arr1= np.array([1,2,3,4,5], float)
arr2 = np.array([5,2,6], float)
sum=arr1+arr2
print("Sum of arr1 and arr2 :",sum)
```

** 2) Subtraction **

```python
arr1= np.array([1,2,3,4,5], int)
arr2 = np.array([5,2,6,7,8], int)
diff=arr1-arr2
print("Sum of arr1 and arr2 :",diff)
```

** 'Floor','ceil' and 'rint' functions **

```python
a = np.array([1.4, 1.7, 1.2], float) 
print("Floor value :",np.floor(a))                        #floor gives the lower integer

print("Ceiling value :",np.ceil(a))                      #ceil gives the upper integer

print("Rounded integer value :",np.rint(a))              #rint gives the rounded integer
```

** 3) linspace(start,stop,num,endpoint,retstep,dtype)**

```python
import numpy as np 

arr1 = np.linspace(10,20,5) 
#start=10,stop=20,num=5,endpoint=true(by default)

#since endpoint=true ,stop value is included i.e 20 will be included in output
print ("Array using linspace :",arr1) 

print('\n')

arr2 = np.linspace(10,20,5,endpoint=False) 

#since endpoint=false ,stop value is not included i.e 20 will not be included in output
print ("Array using linspace and endpoint is false:",arr2)
```

** Slicing of arrays **

**A Python slice object is constructed by giving start, stop, and step parameters 
to the built-in slice function.**
**This slice object is passed to the array to extract a part of array.**

```python
#here we will see some examples of slicing the arrays
#Example 1

import numpy as np 
a = np.arange(10) 
b = a[3:9:2] 
print ("Arrays after slicing :",b)
```

```python
#Example 2

import numpy as np 
arr = np.array([[1,2,3],[3,4,5],[4,5,6]]) 
print('Original Array')
print (arr)  

print('\n')
# slice items starting from index
print ('Now we will slice the array from the index a[2:]')
print (arr[2:])
```

```python
# Example 3
# Here we make use of three dots (...) also known as ellipsis
#If the ellipsis is at rows position we will get the rows as below an so on

# array to begin with 
import numpy as np 
arr = np.array([[1,2,3],[3,4,5],[4,5,6]]) 

print ('Our array is:') 
print (arr) 
print ('\n') 

# this returns array of items in the second column 
print ('The items in the second column are:' ) 
print (arr[...,1])
print ('\n')   

# Now we will slice all items from the second row 
print ('The items in the second row are:')
print (arr[1,...]) 
print ('\n') 

# Now we will slice all items from column 1 onwards 
print ('The items column 1 onwards are:')
print (arr[...,1:])
```

** Array Indexing **

** There are two types of Array Indexing**


**1) Integer Indexing**
**2) Boolean Indexing **

```python
#Integer Indexing
#it involves selecting elements of an array based on its position
#we can get a sub-array of an existing array by using Integer Indexing

import numpy as np 

arr1 = np.array([['a','b'], ['c','d'], ['e','f']])
print("Original array is :")
print(arr1)
print('\n')

sub_arr = arr1[[0,1,2], [0,1,0]] 
print ("Sub-array is :",sub_arr)
```

```python
#Boolean Indexing
#it invovles retrieving elements based on boolean operators
#It is user-friendly way of indexing as the computational statements are easily readable by humans

import numpy as np 
x = np.array([10,20,30,40,50,60,70]) 

# Now we will print the items greater than 30 
print ('The items greater than 30 are:' )
print (x[x > 30])
```

** Broadcasting**


**NumPy has powers to perform mathematical operations even if the arrays are not of similar sizes.**
**The smaller sized array is broadcasted into larger sized array to get the desired output**

```python
#here we performed Broadcasting on two arrays
arr1 = np.array([[0,0],[20,20],[30,30],[40,40]]) # bigger array
arr2 = np.array([5,5])                           #smaller array
   
print ('Array 1 + Array 2') 
print (arr1 + arr2)                              #smaller array is broadcasted and we got the output as below
```

** Iterating over Array**

```python
#here we iterate through every element of an array using for loop

arr1=np.arange(10,110,10)
arr1=arr1.reshape(5,2)

print("Array is :",arr1)

print('\n')

print('Iterated array is :')
for x in np.nditer(arr1):                #nditer() is used to iterate the whole array
    print(x)
```

**Copy Function**

```python
arr = np.arange(12) #we declare an array arr of 12 elements

print ('Our array is:' )
print (arr) 

print('\n')

print ('Applying id() function:') 
print (id(arr))  

print('\n')

print ('arr is assigned to b:') 
b = arr 
print (b)  

print('\n')

print ('b has same id():' )
print (id(b) ) 

print('\n')

print ('Change shape of b:')
b.shape = (6,2) 
print (b)  

print('\n')

print ('Shape of arr also gets changed:')
print (arr)
```

```python
#We have imported interpolate and integrate from scipy 

import numpy as np
from scipy import interpolate, integrate
varA=np.arange(-1,11)
varB=np.exp(-varA/3.0)
varC=interpolate.interp1d(varA,varB)
print("Value is :",integrate.simps(varB[1:-1],varA[1:-1]))
```

```python
#Example of scipy.interpolation

from scipy.interpolate import interp1d

x = np.linspace(0, 10, num=11, endpoint=True) 
y = np.cos(-x**3/9.0)
f = interp1d(x, y) #plot the values of x and y
f2 = interp1d(x, y, kind='cubic')

xnew = np.linspace(0, 10, num=41, endpoint=True)
import matplotlib.pyplot as plt #here we make use of matplotlib to import the necessary functions
plt.plot(x, y, 'o', xnew, f(xnew), '-', xnew, f2(xnew), '--')
plt.legend(['data', 'line', 'cube'], loc='best')
plt.show() #display the actual chart
```
